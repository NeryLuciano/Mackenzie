```python
import urllib.request  

texto = []

#for line in urllib.request.urlopen('/drive/MyDrive/Livro do Curso - Liderança, Capacidade de Aprender e Resiliência.txt'): 
 #  texto.append(line.decode('utf-8'))

 
f = open('Livro do Curso - Liderança, Capacidade de Aprender e Resiliência.txt','r')    # para arquivos locais
for line in f:
   texto.append(line)

for i in range(len(texto)):
    texto[i] = texto[i].lower() # para unicidade
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('!','')
    texto[i] = texto[i].replace('\'','') # elimina ' e "
```

```python
mydict = {}                             # Criando um dicionário vazio

for line in texto:

  line = line.lower()                   # converte para lower
  words = line.split()                  # separa cada palavra
#  print(words)

  for word in words:                                        # para cada palavra em words
    if len(word) > 4 and word not in mydict.keys():         # se palavra não está no dicionário 
      mydict[word] = 1                                      # acrescenta a word com o valor 1
    if len(word) > 4 and word in mydict.keys():             # se a palavra tiver mais que 4 letras
         mydict[word] =  mydict[word] + 1                   # apenas soma 1 ao valor já existente


print(mydict)    
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 13] # somente termos com mais de 13 ocorrências


plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```
